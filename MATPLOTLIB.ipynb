# Anatomy of a Figure

Matplotlib graphs your data on **Figures** (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more **Axes**, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.).

Below mentioned are the parts of Figure:

1. Figure - The whole Figure.

2. Axes - An **Axes** is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects (**be aware of the difference between Axes and Axis**) that provide ticks and tick labels to provide scales for the data in the Axes. Each Axes also has a title (set via set_title()), an x-label (set via **set_xlabel()**), and a y-label set via **set_ylabel()**)

3. Artist - Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects). This includes Text objects, Line2D objects, collections objects, Patch objects, etc. When the Figure is rendered, all of the Artists are drawn to the canvas. Most Artists are tied to an Axes; such an Artist cannot be shared by multiple Axes, or moved from one to another.

# Labelling and Annotating Subplots

```python
# SYNTAX 1
# Initializing a figure (without axes)
fig = plt.figure(figsize=(3, 3), facecolor='lightskyblue')
fig.suptitle("Figure Title")

# Adding Axes (Generates 1 axes or subplot)
ax = fig.subplots()
ax.set_title("Axes Title", loc="left", fontsize="small")

# Adding Annotations
ax.annotate("My Text", xytext=(0.2, 0.2), ha="center",
            xy=(0.5, 0.5), arrowprops={})

# Display the Figure
plt.show()
```

```python
# SYNTAX 2
# Initializing a figure (with 1 axes)
fig, ax = plt.subplots(figsize=(3, 3), facecolor='lightskyblue')
fig.suptitle("Figure Title")

# Adding Axes Title
ax.set_title("Axes Title", loc="left", fontsize="small")

# Adding Axes Grid
ax.grid(True)

# Adding Axes Color
ax.set_facecolor("yellow")

# Adding Annotations
ax.annotate("My Text", xytext=(0.2, 0.2), ha="center",
            xy=(0.5, 0.5), arrowprops={})

# Display the Figure
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout='constrained')
fig.suptitle("Comparision")

data = [data1, data1**2, data1**3]
labels = ['linear', 'quadratic', 'cubic']
colors = ['green', 'orange', 'blue']

for idx, ax in enumerate(axs.ravel()):
    ax.plot(data[0],data[idx], 
            label=labels[idx], 
            color=colors[idx])  # Plot some data on the axes
    
    # Add a horizontal and vertical line
    ax.axhline(1, ls='--', color='red')
    ax.axvline(1, ls='--', color='black')
    
    ax.set_title("Simple Plot")  # Add a title to the axes
    ax.set_xlabel('x label')  # Add an x-label to the axes
    ax.set_ylabel('y label')  # Add a y-label to the axes
        
    ax.legend()  # Add a legend

plt.show()
```

# Using Helper Methods

```python
def my_plotter(axes, data1, data2, kind):
    """
    A helper function to make a graph.
    """
    if kind=="line":
        out = axes.plot(data1, data2)
    else:
        out = axes.scatter(data1, data2)
    return out

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))

my_plotter(ax1, data1, data2, kind="line")
my_plotter(ax2, data3, data4, kind="scatter")

plt.show()
```

# Basic Matplotlib Plots

```python
# Make 4 random data sets

np.random.seed(2)
data1 = np.linspace(0, 100, 10000)
data2 = np.random.randint(1, 6, size=(10000, ))
data3, data4 = np.random.randn(2, 10000)

df = pd.DataFrame({"col_1": data1, 
                   "col_2": data2, 
                   "col_3": data3, 
                   "col_4": data4})

df.head()
```

# Univariate Plotting For Numerical Columns - Histogram and Box Plot

```python
fig, axs = plt.subplots(1, 2, figsize=(6, 3), layout="constrained")
fig.suptitle("Univariate Plotting - Numerical Features")

axs[0].hist(df["col_3"])
axs[0].set_title("Histogram Plot", fontsize="medium")

axs[1].boxplot(df["col_3"])
axs[1].set_title("Box Plot", fontsize="medium")

plt.show()
```

# Univariate Plotting For Categorical Columns - Bar Plot and Pie Chart

```python
fig, axs = plt.subplots(1, 2, figsize=(6, 3), layout="constrained")
fig.suptitle("Univariate Plotting - Categorical Features")

bar_plt_data = df["col_2"].value_counts()
axs[0].bar(bar_plt_data.index, bar_plt_data.values)
axs[0].set_title("Bar Plot", fontsize="medium")

bar_plt_data_normalized = df["col_2"].value_counts(normalize=True)
axs[1].pie(bar_plt_data_normalized.values, labels=bar_plt_data_normalized.index, autopct='%1.1f%%')
axs[1].set_title("Pie Plot", fontsize="medium")

plt.show()
```

# Bivariate Plotting For Num vs Num - Scatter Plot and Hexbin Plot

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout="constrained")
fig.suptitle("Bivariate Plotting - Categorical Features")

axs[0].scatter(df["col_3"], df["col_4"])
axs[0].set_title("Scatter Plot", fontsize="medium")

hb1 = axs[1].hexbin(df["col_3"], df["col_4"])
axs[1].set_title("Hexbin Plot", fontsize="medium")
fig.colorbar(hb1, ax=axs[1])

hb2 = axs[2].hexbin(df["col_3"], df["col_4"], bins="log")
axs[2].set_title("Hexbin Plot with log Color Scale", fontsize="medium")
fig.colorbar(hb2, ax=axs[2])

plt.show()
```
